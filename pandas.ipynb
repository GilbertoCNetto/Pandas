```python
# Leia um arquivo csv e armazene em um dataframe Pandas;

import pandas as pd
dataset = pd.read_csv('./imdb.csv')
```

```python
# Exiba este dataframe completo;
dataset_df = pd.DataFrame(dataset)
display(dataset_df)
```

```python
# Mostre as 5 primeiras linhas e as 5 ultimas linhas do dataframe;
display(dataset_df.head(5))
display(dataset_df.tail(5))
```

```python
# Mostre a transposta deste dataframe
dataset_df.T
```

```python
# Mostre os valores de uma Ãºnica coluna e tambÃ©m a quantidade que cada valor aparece;
display(dataset_df['rating'].unique())
display(dataset_df['rating'].value_counts())
```

```python
# Mostre o dataframe agrupado por uma coluna escolhida;
df_agrupado = dataset_df.groupby('votes').agg('sum').sort_values('lister-item-index')
display(df_agrupado)
```

```python
# Selecione uma coluna e dentro desta trÃªs valores especÃ­ficos para trocar por um NaN;

# Exibindo o DataFrame original
print("DATAFRAME ORIGINAL:")
display(dataset)

# Selecionando a coluna e substituindo valores especÃ­ficos por NaN
values_to_replace = [1.0, 3.0, 5.0]
column_name = 'lister-item-index'
dataset_alterado = dataset_df
dataset_alterado[column_name] = dataset_df[column_name].replace(values_to_replace, pd.NA)

# Exibindo o DataFrame com os valores substituÃ­dos
print("\nDATA FRAME COM VALORES ALTERADOS:")
display(dataset_alterado)
```

```python
# Mostre o dataframe sem os registros que possuem NaN;

# Exibindo o DataFrame Alterado
print("DataFrame Alterado:")
display(dataset_alterado)

# Removendo os registros que possuem valor NaN
dataset_clean = dataset_alterado.dropna()

# Exibindo o DataFrame com os registros que possuem NaN
print("\nDataFrame sem registros NaN:")
display(dataset_clean)
```

```python
# Mostre os valores estatÃ­sticos do dataframe;

# Selecionando as colunas desejadas
selected_columns = ['runtime', 'votes']
selected_data = dataset[selected_columns]
another_columns = ['lister-item-index', 'rating']
selected_data2 = dataset[another_columns]
# Exibindo os valores estatÃ­sticos das colunas selecionadas
print("Valores estatÃ­sticos das colunas selecionadas:")
display(selected_data.describe())
display(selected_data2.describe())
```

```python
# Crie os seguintes grÃ¡ficos baseados no dataframe: histograma, dispersÃ£o e colunas

import matplotlib.pyplot as plt

# Histograma
plt.figure(figsize=(8, 6))
plt.hist(dataset['rating'], bins=10, color='aquamarine', edgecolor='black')
plt.title('DistribuiÃ§Ã£o de Notas x FrequÃªncias')
plt.xlabel('Notas')
plt.ylabel('FrequÃªncia')
# FrequÃªncia Ã© o nÃºmero de vezes que aquele "Rating"(nota) aparece
plt.grid(False)
plt.show()

# GrÃ¡fico de dispersÃ£o (Notas vs. PosiÃ§Ã£o)
plt.figure(figsize=(8, 6))
plt.scatter(dataset.head(20)['rating'], dataset.head(20)['lister-item-index'], color='dodgerblue', alpha=1)
plt.title('Notas vs. PosiÃ§Ã£o: Melhores 20 SÃ©ries')
plt.xlabel('Notas')
plt.ylabel('PosiÃ§Ã£o')
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()

# GrÃ¡fico de barras para a posiÃ§Ã£o
plt.figure(figsize=(8, 6))
plt.bar(dataset.head(15)['lister-item-index'], dataset.head(15)['rating'], color='crimson', alpha=1)
plt.title('Notas por PosiÃ§Ã£o')
plt.xlabel('PosiÃ§Ã£o')
plt.ylabel('Notas')
plt.yticks(range(1, 11))
plt.xticks(dataset.head(15)['lister-item-index'])
plt.grid(axis='y')
plt.show()
```
